```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
np.random.seed(42)
num_samples = 100
```

```python
class_0_x = np.random.normal(2, 1, num_samples)  
class_0_y = np.random.normal(2, 1, num_samples)
```

```python
class_1_x = np.random.normal(6, 1, num_samples)
class_1_y = np.random.normal(6, 1, num_samples)
```

```python
X = np.concatenate((np.column_stack((class_0_x, class_0_y)), np.column_stack((class_1_x, class_1_y))))
y = np.concatenate((np.zeros(num_samples), np.ones(num_samples)))
```

```python
log_reg = LogisticRegression()
log_reg.fit(X, y)
```

```python
plt.scatter(class_0_x, class_0_y, color='red', label='Class 0')
plt.scatter(class_1_x, class_1_y, color='blue', label='Class 1')
```

```python
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = log_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3, levels=[0, 0.5, 1], colors=['red', 'blue'])
```

```python
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Example')
plt.legend()
plt.grid(True)
plt.show()
```
