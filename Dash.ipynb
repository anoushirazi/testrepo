```python
import dash
from dash import dcc, html
import pandas as pd
import plotly.express as px
from dash.dependencies import Input, Output

# Initialize the Dash application
app = dash.Dash(__name__)

# Load data
data = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv')

# Filter or prepare recession data as needed
recession_data = data  # Modify this line to filter or prepare data for recession statistics

# Define your Dash app layout here
app.layout = html.Div([
    # Assuming you need a dropdown to select the statistics type
    dcc.Dropdown(
        id='dropdown-statistics',
        options=[
            {'label': 'Recession Period Statistics', 'value': 'Recession Period Statistics'},
            # Add more options as needed
        ],
        value='Recession Period Statistics',  # Default value or could be None
    ),
    html.Div(id='output-container', className='chart-grid', style={'display': 'flex'}),
])

# Define the callback for updating the output container with plots
@app.callback(
    Output('output-container', 'children'),
    [Input('dropdown-statistics', 'value')]
)
def update_output_container(selected_statistic):
    if selected_statistic == 'Recession Period Statistics':
        # Plot 1: Line Chart
        yearly_rec = recession_data.groupby('Year')['Automobile_Sales'].mean().reset_index()
        R_chart1 = dcc.Graph(
            figure=px.line(yearly_rec, x='Year', y='Automobile_Sales', title="Automobile Sales Fluctuation During Recession Years"))

        # Plot 2: Bar Chart
        vehicle_type_avg = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
        R_chart2 = dcc.Graph(
            figure=px.bar(vehicle_type_avg, x='Vehicle_Type', y='Automobile_Sales', title="Average Sales by Vehicle Type"))

        # Plot 3: Pie Chart
        expenditure_share = recession_data.groupby('Vehicle_Type')['Expenditure'].sum().reset_index()
        R_chart3 = dcc.Graph(
            figure=px.pie(expenditure_share, values='Expenditure', names='Vehicle_Type', title="Expenditure Share by Vehicle Type"))

        # Plot 4: Bar Chart
        effect_unemployment = recession_data.groupby(['Vehicle_Type', 'Unemployment_Rate'])['Automobile_Sales'].mean().reset_index()
        R_chart4 = dcc.Graph(
            figure=px.bar(effect_unemployment, x='Vehicle_Type', y='Automobile_Sales', color='Unemployment_Rate', title="Impact of Unemployment on Vehicle Sales"))

        return [
            html.Div(className='chart-item', children=[R_chart1]),
            html.Div(className='chart-item', children=[R_chart2]),
            html.Div(className='chart-item', children=[R_chart3]),
            html.Div(className='chart-item', children=[R_chart4])
        ]
    else:
        return []

# Run the Dash app server
if __name__ == '__main__':
    app.run_server(debug=True, port=8080)
```
